    Univariate Analysis: Analyze individual variables to understand their distributions

    Bivariate Analysis: Explore relationships between pairs of variables to identify 
        potential correlations or associations.

    Multivariate Analysis: Investigate interactions between multiple variables to 
        gain deeper insights into the data.
